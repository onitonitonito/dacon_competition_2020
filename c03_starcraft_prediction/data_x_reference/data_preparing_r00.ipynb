```python
def compare_columns(column_1, column_2, echo=True):
    """
    # compare_columns(column_1, column_2, echo=True)
    # return bool-array of two columns
    """
    cols_1, cols_2 = list(column_1), list(column_2)
    (len_c1, len_c2) = (len(cols_1), len(cols_2))

    delta = 0 if len_c1 == len_c2 else len_c1 - len_c2

    echo_lines = [
            "NUM  COLUMN-01" + " "*16 + "COLUMN-02" + " "*16 + "IS_SAME",
            "--"*32,
        ]

    if abs(delta) > 0:
        _extend = ['-'] * abs(delta)

        if delta > 0:
            cols_2.extend(_extend)
        else:
            cols_1.extend(_extend) 

    for i, (c1, c2) in enumerate(zip(cols_1, cols_2)):
        _line = f"{i:02} | {c1:23} | {c2:23} | {c1 == c2}"
        echo_lines.append(_line) 
    
    if echo:
        _lines = "\n".join(echo_lines)
        print(_lines)
    return cols_1 == cols_2    


compare_columns(X_train.columns[:15], X_test.columns, echo=1)
```
