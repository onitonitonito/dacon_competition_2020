## Pandas Merge, Join, Concatenate
> - Merge, join, and concatenate — pandas 1.0.3 documentation
> - https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html


```python
df1 = pd.DataFrame({'A': ['A0', np.NaN, 'A1', 'A7', 'A0'],
                    'B': ['B0', 'B3', np.NaN, 'B12','B9'],
                    'C': ['C0', 'C3', 'C9', 'C10', 'C11'],
                    'D': ['D8', np.NaN, 'D2', 'D3', 'D7']},
                   index=np.arange(5))


df2 = pd.DataFrame({'A': [np.NaN, 'A6',],
                    'B': ['B4', 'B6',],
                    'C': ['C5', 'C7'],
                    'D': ['D5', '',]},
                   index=[12, 7])


df3 = pd.DataFrame({'A': ['A8', 'A4', 'A5',np.NaN,],
                    'B': ['B8',  np.NaN,'B6','B12'],
                    'C': ['C10',  'C12','C0','C5'],
                    'D': ['D10', 'D5', 'D11','D1']},
                   index=np.arange(10, 14))
```

```python
"""
# df1 + df2 + df3 = concatination!
# df_1 = df_1.fillna('-')
"""

df_0 = pd.concat([df1, df2, df3], axis=1, join='outer')
df_0
```

```python
"""
# 칼럼 맨 앞열에 삽입하기
# df_1['old_index'] = df_1.index    # placed last
# python - how do I insert a column at a specific column index in pandas? 
# - Stack Overflow = https://bit.ly/2Rq1CEU
"""

pos_idx = 0
column = 'old_index'
value = df_0.index

df_0.insert(pos_idx, column=column, value=value)
df_0
```
